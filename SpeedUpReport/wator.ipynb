```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = {
    'one': [3.70, 3.81, 3.71, 3.70, 3.72, 3.70, 3.71, 3.70, 3.70],
    'two': [3.59, 3.60, 3.62, 3.60, 3.59, 3.59, 3.60, 3.60, 3.60],
    'four': [3.93, 3.90, 3.85, 3.94, 3.91, 3.90, 3.90, 3.91, 3.91],
    'eight': [3.52, 3.53, 3.48, 3.49, 3.49, 3.49, 3.50, 3.50, 3.49]
}

# Convert seconds to milliseconds and calculate averages
averages = {key: np.mean(np.array(value) * 1000) for key, value in data.items()}

# Create bar chart with different colors
colors = ['skyblue', 'salmon', 'lightgreen', 'lightcoral']

plt.figure(figsize=(10, 6))
bars = plt.bar(averages.keys(), averages.values(), color=colors)

# Add labels and title
plt.title('Average Execution Time per Thread Count')
plt.xlabel('Number of Threads')
plt.ylabel('Average Execution Time (ms)')

# Set y-axis increments to 250
plt.yticks(np.arange(0, max(averages.values()) + 250, 250))

# Add data labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10, round(yval, 2), ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = {
    'one': [3.70, 3.81, 3.71, 3.70, 3.72, 3.70, 3.71, 3.70, 3.70],
    'two': [3.59, 3.60, 3.62, 3.60, 3.59, 3.59, 3.60, 3.60, 3.60],
    'four': [3.93, 3.90, 3.85, 3.94, 3.91, 3.90, 3.90, 3.91, 3.91],
    'eight': [3.52, 3.53, 3.48, 3.49, 3.49, 3.49, 3.50, 3.50, 3.49]
}

# Convert seconds to milliseconds
data_ms = {key: np.array(value) * 1000 for key, value in data.items()}

# Plot line chart
plt.figure(figsize=(10, 6))
for key, value in data_ms.items():
    plt.plot(value, label=f'{key} Threads', marker='o')

# Add labels and title
plt.title('Execution Time Trends for Different Thread Counts')
plt.xlabel('Run Number')
plt.ylabel('Execution Time (ms)')

# Add legend
plt.legend()

# Show grid
plt.grid(True)

plt.show()
```
